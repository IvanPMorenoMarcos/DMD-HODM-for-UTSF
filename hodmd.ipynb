```python
# import all dependencies
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12,7)
import numpy as np
from pydmd import HODMD
```

```python
# import files from google drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
# load and plot training dataset
training = pd.read_csv('/content/drive/My Drive/Dynamic-Mode-Decomposition/airline_passengers_training.csv', index_col=0, header=None, parse_dates=True, squeeze=True)
training.plot()
plt.show()
```

```python
# create and fit a HODMD model
hodmd = HODMD(svd_rank=0,
          opt=True,
          exact=True,
          rescale_mode=None,
          forward_backward=False,
		  sorted_eigs=False,
          tlsq_rank=0,
          d=50)
hodmd.fit(training.values)
```

```python
# plot predictions vs actuals on training dataset
plt.plot(training.values, label='Actuals')
plt.plot(hodmd.reconstructed_data[0], label='Predictions')
plt.legend()
plt.show()
```
